# Dataset

```python
drive.mount('/content/drive')
path = Path('/content/drive/MyDrive/......./test_second_dataset') # change depending on the location of the database
```

```python
def createFileList(myDir, format='.jpg'):
    fileList = []
    print(myDir)
    for root, dirs, files in os.walk(myDir, topdown=False):
        for name in files:
            if name.endswith(format):
                fullName = os.path.join(root, name)
                fileList.append(fullName)
    return fileList
myFileList = createFileList(path)
```

```python
df=pd.DataFrame({ 'pathFiles' : [], 'label':[]})
```

```python
ruta=myFileList
label=[]
for i in range(len(myFileList)):
    label.append(myFileList[i].split(os.sep)[:][-2])
df['pathFiles']=ruta
df['label']=label
df.head()
```
